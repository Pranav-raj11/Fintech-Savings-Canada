## Cleaning 'Biggest Banks' Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("./Datasets/Biggest Banks (csv).csv")
df
```

```python
dfe = pd.read_excel("./Datasets/Biggest Banks (excel).xlsx")
dfe
```

```python
df.drop_duplicates()
```

```python
df["Value"] = df["Value"].str.strip("C$")
df.head(40)
```

```python
for i in df.index:
    val = df.loc[i, "Value"]
    if pd.isna(val):
        continue
    val = str(val).strip()
    if val.endswith('B'):
        df.loc[i, "Value"] = float(val[:-1]) * 1e9
    if val.endswith('T'):
        df.loc[i, "Value"] = float(val[:-1]) * 1e12

df.head(40)
```

```python
df.tail(50)
```

```python
df["Value"] = pd.to_numeric(df["Value"], errors="coerce").round(0)
df.head(40)
```

```python
df.to_csv("cleaned_biggest_bank_data.csv")
```

```python
df.to_csv("cleaned_bank_data.csv", index=False, float_format="%.2f")
```
